```python
# Import core libraries
import numpy as np                  # Numerical computations
import matplotlib.pyplot as plt     # Plotting library
```

```python
# Basic Line Plot
# Create sample data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Plot y vs x
#plt.plot(x)
#plt.plot(y)
plt.plot(x, y)

# Display the plot
plt.show()
```

```python
# Adding Labels, Title, and Grid

plt.plot(x, y)

# Add axis labels
plt.xlabel("X values")
plt.ylabel("Y values")

# Add title
plt.title("Simple Line Plot")

# Add grid
plt.grid(True)

plt.show()
```

```python
# Multiple Lines on Same Plot
x = np.linspace(0, 10, 100)

#y = np.sin(x)
#plt.plot(x)
#plt.plot(y)
plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")

# Add legend
plt.legend()

plt.title("Multiple Lines")
plt.show()
```

```python
# Using Different Line Styles & Colors

plt.plot(x, np.sin(x), 'r--', label="Red Dashed")
plt.plot(x, np.cos(x), 'b-', label="Blue Solid")

plt.legend()
plt.show()
```

```python
# Scatter Plot

x = np.linspace(0, 10, 30)
y = np.sin(x)

# Scatter plot using plot()
plt.plot(x, y, 'o')

plt.title("Scatter Plot")
plt.show()
```

```python
# Histogram
# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30)

plt.title("Histogram")
plt.show()
```

```python
# Vertical Bar Chart
values = [5, 25, 50, 20]

plt.bar(range(len(values)), values)

plt.title("Bar Chart")
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(range(len(values)), values)

plt.title("Horizontal Bar Chart")
plt.show()
```

```python
# Stacked Bar Chart
A = [15, 30, 45, 22]
B = [15, 25, 50, 20]

x = range(4)

plt.bar(x, A, label="A")
plt.bar(x, B, bottom=A, label="B")

plt.legend()
plt.title("Stacked Bar Chart")
plt.show()
```
