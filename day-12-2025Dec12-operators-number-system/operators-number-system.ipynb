## Swap 2 variables

```python
a, b = 5, 6
print(a)
print(b)
```

```python
temp = a
a = b
b = temp
print(a)
print(b)
```

```python
a = a + b
b = a - b
a = a - b
print(a)
print(b)
```

```python
a, b = b, a
print(a)
print(b)
```
