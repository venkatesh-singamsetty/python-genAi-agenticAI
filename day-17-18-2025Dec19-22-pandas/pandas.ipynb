```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = 6, 3

import warnings
warnings.filterwarnings('ignore')
```

```python
vis1 = sns.distplot(df['InternetUsers'])  # univariate analysis i.e using one variable
plt.show()
```

```python
vis2 = sns.displot(df["InternetUsers"])
plt.show(vis2)
```

```python
vis3 = sns.distplot(df["InternetUsers"], bins=15)
plt.show(vis3)
```

```python
vis4 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate') 
plt.show(vis4)
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate')
plt.show(vis5) 
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate', fit_reg=False)
plt.show(vis5)
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate', fit_reg=True)
plt.show(vis5) 
```

```python
vis8 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color

plt.show(vis8)
```
