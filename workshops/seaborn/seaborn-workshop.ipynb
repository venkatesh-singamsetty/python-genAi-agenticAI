```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
```

```python
# 1. scatterplot
# data=tips → Pandas DataFrame
# x="total_bill" → Total bill amount
# y="tip" → Tip amount
# hue="time" → Color-coded by Lunch vs Dinner
# size="size" → Marker size represents party size
# palette="deep" → High-contrast, professional color palette

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", palette="deep")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# 2. lineplot
# data=tips → Pandas DataFrame
# x="size" → Party size (number of people)
# y="total_bill" → Total bill amount
# hue="sex" → Separate line for Male and Female
# marker='o' → Places a circle marker at each aggregated point

sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", marker='o')
plt.title("Line Plot of Total Bill vs Size")
plt.show()
```

```python
# ci=None → Removes confidence interval shading

sns.lineplot(data=tips, x="size", y="total_bill", hue="sex", ci=None, marker='o')
plt.title("Line Plot of Total Bill vs Size")
plt.show()
```

```python
# 3. barplot
# data=tips → Pandas DataFrame
# x='day' → Categorical variable (Thu, Fri, Sat, Sun)
# y='total_bill' → Numeric variable
# hue='sex' → Separate bars for Male and Female
# palette='muted' → Soft, professional color palette

sns.barplot(data=tips, x='day', y='total_bill', hue='sex', palette='muted')
plt.title("Bar Plot of Total Bill by Day")
plt.show()
```

```python
tips.columns
```

```python
# 4. Box plot
# data=tips → Pandas DataFrame
# x="day" → Categorical variable (Thu, Fri, Sat, Sun)
# y="tip" → Numeric variable (tip amount)
# hue="smoker" → Splits boxes into Smoker and Non-Smoker groups
# palette="Set2" → Color-blind friendly categorical palette

sns.boxplot(data=tips, x="day", y="tip", hue="smoker", palette="Set2")
plt.title("Box Plot of Tip by Day and Smoker Status")
plt.show()
```

```python
# 5. violin plot
# data=tips → Pandas DataFrame
# x="day" → Categorical variable (Thu, Fri, Sat, Sun)
# y="total_bill" → Numeric variable
# hue="time" → Splits data by Lunch vs Dinner
# split=True → Shows both categories inside one violin (left/right)
# palette="pastel" → Soft, light colors for readability

sns.violinplot(data=tips, x="day", y="total_bill", hue="time", split=True, palette="pastel")
plt.title("Violin Plot of Total Bill by Day and Time")
plt.show()
```

```python
tips.columns
```

```python
# 6. countplot
# data=tips → Pandas DataFrame
# x="day" → Categorical variable (Thu, Fri, Sat, Sun)
# hue="smoker" → Splits counts into Yes / No smoker groups
# palette="dark" → Dark, high-contrast color palette

sns.countplot(data=tips, x="day", hue='smoker', palette="dark")
plt.title("Count Plot of Days by Smoker Status")
plt.show()
```

```python
# 7. regression plot
# data=tips → Pandas DataFrame
# x="total_bill" → Independent variable
# y="tip" → Dependent variable
# scatter_kws={'s': 50} → Sets marker size for data points
# line_kws={'color': 'red'} → Styles the regression line

sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'s':50}, line_kws={'color': 'red'})
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

```python
# 8. histogram
# data=tips → Pandas DataFrame
# x="total_bill" → Numeric variable to analyze
# bins=20 → Number of bars (controls granularity)
# kde=True → Adds a Kernel Density Estimate curve
# color="blue" → Sets bar and KDE color

sns.histplot(data=tips, x="total_bill", bins=20, kde=True, color="blue")
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

```python
# 9. pairplot
# tips → Pandas DataFrame
# vars=[...] → Numeric variables to include
# hue="sex" → Color-codes points by Male vs Female
# palette="husl" → Evenly spaced, high-contrast colors

sns.pairplot(tips, hue="sex", vars=["total_bill", "tip", "size"], palette="husl")
plt.suptitle("Pair Plot: Numeric variables by Gender", y=1.02)
plt.show()
```

```python
# 10. catplot
# data=tips → Pandas DataFrame
# x="day" → Categorical variable (Thu, Fri, Sat, Sun)
# y="tip" → Numeric variable
# hue="sex" → Separate points/lines for Male and Female
# kind="point" → Creates a point plot (mean + CI)
# palette="bright" → High-contrast colors

sns.catplot(data=tips, x="day", y="tip", hue="sex", kind="point", palette="bright")
plt.title("Cat Plot: Tips by day and gender")
plt.show()
```

```python
# 11. jointplot
# data=tips → Pandas DataFrame
# x="total_bill" → X-axis variable
# y="tip" → Y-axis variable
# kind="scatter" → Scatter plot in the joint area
# hue="smoker" → Color-coded by Smoker vs Non-Smoker
# palette="coolwarm" → Color gradient for hue categories
# color="purple" → Base color (largely ignored when hue is set)

sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="smoker", color="purple", palette="coolwarm")
plt.suptitle("Joint Plot: Total Bill vs Tip", y=1.02)
plt.show()
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="smoker", palette="coolwarm")
plt.suptitle("Joint Plot: Total Bill vs Tip", y=1.02)
plt.show()
```

```python
# 12. Facetgrid
# sns.FacetGrid(...)
## Creates a grid of subplots based on categorical variables.
## col='time' → Separate columns for Lunch vs Dinner
## row='smoker' → Separate rows for Smoker vs Non-Smoker
## margin_titles=True → Moves titles to the margins for clarity

# .map(sns.histplot, 'total_bill', bins=20, kde=True)
## Plots a histogram of total_bill in each facet
## bins=20 → Controls number of bins
## kde=True → Adds a smoothed density curve

# .add_legend()
## Adds a legend to explain colors or hues (if used)

g = sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', bins=20, kde=True).add_legend()
g
```

```python
# 13. stripplot
# data=tips → Pandas DataFrame
# x='day' → Categorical variable (Thu, Fri, Sat, Sun)
# y='tip' → Numeric variable
# hue='sex' → Color-coded by Male / Female
# jitter=True → Adds horizontal noise so overlapping points are spread out
# palette='Set1' → High-contrast, readable colors

sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("strip plot: Tips by data and gender")
plt.show()
```

```python
# 14. KDE plot
# data=tips → Pandas DataFrame
# x='total_bill' → Numeric variable to plot
# hue='sex' → Splits the KDE by Male / Female
# fill=True → Fills the area under the curve for better visualization
# palette='tab10' → Distinct, professional colors

sns.kdeplot(data=tips, x='total_bill',hue='sex', fill=True, palette='tab10')
plt.title("kde plot:Total bill density by gender")
plt.show()
```
