# print

```python
print(10) #- 1 arg 
print(10,20) # 2 arg
print('python') # string arg
print(10,20,'python') # 3 arg
```

```python
num1, num2 = 20, 30

add=num1+num2

print('The addition of',num1,'and',num2,'is=',add)
```

```python
num1, num2, num3 =10, 20, 30
add = num1 + num2 + num3

#format string var = {}
print('The addition of {} and {} and {} is= {}'.format(num1, num2, num3, add))

#f string var within braces
print(f'The addition of {num1} and {num2} and {num3} is= {add}')
      
```
