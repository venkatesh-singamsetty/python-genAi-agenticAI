## set operation

```python
s3 = {1,2,3,4,5}
s4 = {4,5,6,7,8}
s5 = {8,9,10}
```

```python
s3.union(s4)
```

```python
s3 | s4 | s5
```

```python
s5.union(s3, s4)
```

```python
print(s3)
print(s4)
print(s5)
```

```python
s3.intersection(s4)
```

```python
s4 & s5
```

```python
s3 & s4 & s5
```

```python
s3.intersection(s4, s5)
```

```python
print(s3)
print(s4)
print(s5)
```

```python
s3.difference(s4)
```

```python
s4 - s3
```

```python
s3 - s5
```

```python
s5 - s3
```

```python
print(s3)
print(s4)
print(s5)
```

```python
s3.symmetric_difference(s4)
```

```python
s4 ^ s5
```

```python
len(s3)
```

```python
print(s3)
print(s4)
print(s5)
```

```python
s3.symmetric_difference_update(s4)
s3
```
